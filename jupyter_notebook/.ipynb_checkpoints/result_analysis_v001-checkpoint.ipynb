```python
dementia_data = []
control_data = []

with open('../../data/dementia.txt', 'r') as f:
    dementia_raw = f.readlines()
    for i in range(len(dementia_raw)):
        if i%2==0:
            dementia_data.append(dementia_raw[i+1])
with open('../../data/control_51.txt', 'r') as f:
    control_raw = f.readlines()
    for i in range(len(control_raw)):
        if i%2==0:
            control_data.append(control_raw[i+1])
```
